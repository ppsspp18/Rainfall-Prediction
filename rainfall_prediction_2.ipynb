```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('weatherAUS.csv')
df.head(15)
```

```python
# one hot encoding
categorical_columns = ['RainToday', 'WindGustDir', 'WindDir9am', 'WindDir3pm']
df1 = pd.get_dummies(data=df, columns=categorical_columns)


df1.replace(['No', 'Yes'], [0, 1], inplace=True)
df1.drop(['Date', 'Location'], axis=1, inplace=True)
df1 = df1.astype(float)
df1.dropna(inplace=True)
X = df1.drop(columns='RainTomorrow', axis=1)
Y = df1['RainTomorrow']

print(df1.head(15))
print(df1.columns)
print(df1.isna().sum())
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler


scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

pca = PCA(n_components=0.95)
X_pca = pca.fit_transform(X_scaled)

print("Shape of PCA-transformed data:", X_pca.shape)
```

```python

print("Shape of PCA-transformed data before reshaping:", X_pca.shape)

n_samples = X_pca.shape[0]
n_features = X_pca.shape[1]

n_timesteps = 1
X_reshaped = X_pca.reshape(n_samples, n_timesteps, n_features)

print("Shape of reshaped data for LSTM:", X_reshaped.shape)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

n_timesteps = X_reshaped.shape[1]
n_features = X_reshaped.shape[2]

num_classes = 1

model = Sequential()
model.add(LSTM(units=50, input_shape=(n_timesteps, n_features)))
model.add(Dense(units=num_classes, activation='sigmoid'))

model.summary()
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model.fit(X_reshaped, Y, epochs=50, batch_size=32)
```

```python
from sklearn.metrics import f1_score

loss, accuracy = model.evaluate(X_reshaped, Y)
print(f'Loss: {loss:.4f}')
print(f'Accuracy: {accuracy:.4f}')

y_pred_prob = model.predict(X_reshaped)

y_pred = (y_pred_prob > 0.5).astype(int)

f1_lstm = f1_score(Y, y_pred)
print(f'F1 Score: {f1:.4f}')
```

```python
from tensorflow.keras.layers import GRU

gru_model = Sequential()
gru_model.add(GRU(units=50, input_shape=(n_timesteps, n_features)))
gru_model.add(Dense(units=num_classes, activation='sigmoid'))

gru_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

gru_model.fit(X_reshaped, Y, epochs=200, batch_size=32)
```

```python
loss_gru, accuracy_gru = gru_model.evaluate(X_reshaped, Y)

print(f'GRU Model Loss: {loss_gru:.4f}')
print(f'GRU Model Accuracy: {accuracy_gru:.4f}')

y_pred_prob = gru_model.predict(X_reshaped)

y_pred = (y_pred_prob > 0.5).astype(int)

f1_gru = f1_score(Y, y_pred)
print(f'F1 Score: {f1_gru:.4f}')
```

```python
from tensorflow.keras.layers import SimpleRNN

simplernn_model = Sequential()
simplernn_model.add(SimpleRNN(units=50, input_shape=(n_timesteps, n_features)))
simplernn_model.add(Dense(units=num_classes, activation='sigmoid'))

simplernn_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

print("Training SimpleRNN model...")
simplernn_model.fit(X_reshaped, Y, epochs=10, batch_size=32)
print("SimpleRNN model training finished.")
```

```python
loss_simplernn, accuracy_simplernn = simplernn_model.evaluate(X_reshaped, Y)

print(f'SimpleRNN Model Loss: {loss_simplernn:.4f}')
print(f'SimpleRNN Model Accuracy: {accuracy_simplernn:.4f}')

y_pred_prob = simplernn_model.predict(X_reshaped)

y_pred = (y_pred_prob > 0.5).astype(int)

f1_srnn = f1_score(Y, y_pred)
print(f'F1 Score: {f1_srnn:.4f}')
```

```python
import pandas as pd

performance_data = {
    'Model': ['LSTM', 'GRU', 'SimpleRNN'],
    'Loss': [loss, loss_gru, loss_simplernn],
    'Accuracy': [accuracy, accuracy_gru, accuracy_simplernn],
    'F1_score' : [f1_lstm,f1_gru,f1_srnn]
}

performance_df = pd.DataFrame(performance_data)
display(performance_df)
```

```python
from sklearn.model_selection import train_test_split

X_train1, X_test1, Y_train1, Y_test1 = train_test_split(
    X, Y, test_size=0.2, random_state=42, stratify=Y
)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled1 = scaler.fit_transform(X_train1)
X_test_scaled1 = scaler.transform(X_test1)

proto_yes = np.mean(X_train_scaled1[Y_train1 == 1], axis=0)
proto_no = np.mean(X_train_scaled1[Y_train1 == 0], axis=0)
```

```python
from scipy.spatial.distance import euclidean

def predict(sample):
    dist_yes = euclidean(sample, proto_yes)
    dist_no = euclidean(sample, proto_no)
    return 1 if dist_yes < dist_no else 0

Y_pred1 = [predict(sample) for sample in X_test_scaled1]
```

```python
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix, classification_report

accuracy = accuracy_score(Y_test1, Y_pred1)
f1_lwp = f1_score(Y_test1, Y_pred1)

print(f"Accuracy: {accuracy:.4f}")
print(f"F1 Score: {f1_lwp:.4f}")

print(classification_report(Y_test1, Y_pred1))
```
